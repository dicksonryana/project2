```python
import pandas as pd
import requests as rq
import lxml as xl
import pytrends
from pytrends.request import TrendReq
import numpy
import datetime
pytrends = TrendReq()
```

```python
topic_ls = ["guns", "germs", "recreation"]
gun_ls = ["ammunition", "gun stores", "ar-15"]
germs_ls = ["hand sanitizer", "bleach", "toilet paper"]
tv_ls = ["netflix", "weed delivery", "beer delivery"]
```

```python
my_req = TrendReq(hl='en-US', tz=360)
my_req.build_payload(topic_ls, timeframe = 'today 5-y')
df_topic = my_req.interest_over_time()
df_topic.head()
```

```python
my_req2 = TrendReq(hl='en-US', tz=360)
my_req2.build_payload(gun_ls, timeframe = 'today 5-y')
df_guns = my_req2.interest_over_time()
df_guns.head()
```

```python
my_req = TrendReq(hl='en-US', tz=360)
my_req.build_payload(germs_ls, timeframe = 'today 5-y')
df_germs = my_req.interest_over_time()
df_germs.head()
```

```python
my_req = TrendReq(hl='en-US', tz=360)
my_req.build_payload(tv_ls, timeframe = 'today 5-y')
df_rec = my_req.interest_over_time()
df_rec.head()
```

```python
df_topic = df_topic.drop("isPartial", axis = 1)
df_guns = df_guns.drop("isPartial", axis = 1)
df_germs = df_germs.drop("isPartial", axis = 1)
df_rec = df_rec.drop("isPartial", axis = 1)
```

```python
df_topic = df_topic.reset_index()
df_guns = df_guns.reset_index()
df_germs = df_germs.reset_index()
df_rec = df_rec.reset_index()
df_rec
```

```python
x = datetime.datetime(2019, 4, 1)
year = []
for index, inner in df_topic.iterrows():
    if inner["date"].year != x.year:
        df_topic = df_topic.drop(index)
    elif inner["date"].month > x.month:
        df_topic = df_topic.drop(index)
df_topic.to_csv("goog_csvs/2019_topic.csv")
```
